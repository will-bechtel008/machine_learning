Imports

```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

Bivariate Functions

```
def f1(x, y):
    return (2*x) ** 2 + (2*y) ** 2 - 4

def f2(x, y):
    return np.cos(x) - np.cos(y)

def f3(x, y):
    return x ** 2 - y ** 2
```

Tests

```
# Function 1
def f1(x, y):
    return (2*x) ** 2 + (2*y) ** 2 - 4

x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 50, cmap = 'binary')

ax.scatter(0,0)
ax.text(0, 3, 0, "p1", color = 'blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Function 1')
ax.view_init(0, 20)
```

```
# Function 2
def f2(x, y):
    return np.cos(x) - np.cos(y)

x = np.linspace(-6, 2, 100)
y = np.linspace(-6, 2, 100)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 50, cmap = 'binary')

ax.scatter(0,-3.2, 2)
ax.text(1, -3, 2.5, "p2", color = 'blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Function 2')
ax.view_init(15, 30)
```

```
# Function 3
def f3(x, y):
    return x ** 2 - y ** 2

x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

X, Y = np.meshgrid(x, y)
Z = f3(X, Y)

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 50, cmap = 'binary')

ax.scatter(0, 0)
ax.text(0, -0.5, -10, "p3", color = 'blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Function 3')
ax.view_init(0, 30)
```
